```python
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
```

```python
class Model(nn.Module):
  def __init__(self):
    super(Model, self).__init__()
    self.fc1 = nn.Linear(48, 64)
    self.fc2 = nn.Linear(64, 32)
    self.fc3 = nn.Linear(32, 10)
    self.fc4 = nn.Linear(10, 2)
    self.dropout1 = nn.Dropout(0.2)
    self.dropout2 = nn.Dropout(0.3)

  def forward(self, x):
    x = F.relu(self.fc1(x))
    x = self.dropout1(x)
    x = F.relu(self.fc2(x))
    x = self.dropout1(x)
    x = F.relu(self.fc3(x))
    x = self.dropout2(x)
    x = self.fc4(x)
    return x
```

```python
batch_size = 128
num_epochs = 50
```

```python
model = Model()
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
X_tensor = torch.tensor(X, dtype=torch.float32)
y_tensor = torch.tensor(y.values, dtype=torch.long)
```

```python
X_train = X_tensor[:int(len(X_tensor)*0.8)]
y_train = y_tensor[:int(len(y_tensor)*0.8)]

X_test = X_tensor[int(len(X_tensor)*0.8):]
y_test = y_tensor[int(len(y_tensor)*0.8):]
```

```python
for epoch in range(num_epochs):
  for i in range(0, len(X_train), batch_size):
    batch_X = X_train[i:i+batch_size]
    batch_y = y_train[i:i+batch_size]

    optimizer.zero_grad()
    outputs = model(batch_X)
    # Apply sigmoid activation to get probabilities for each class
    probs = torch.sigmoid(outputs)
    # Select the probability for the positive class (assuming binary classification)
    probs = probs[:, 1]
    loss = criterion(probs, batch_y.float()) # Convert batch_y to float for BCELoss
    loss.backward()
    optimizer.step()

    def accuracy(X, y):
        correct = 0
        total = len(y)
        model.eval()  # Set the model to evaluation mode

        with torch.no_grad():
            outputs = model(X)
            _, predicted = torch.max(outputs.data, 1)
            correct += (predicted == y).sum().item()

        return (correct / total) * 100

    model.eval()  # Set the model to evaluation mode
    with torch.no_grad():
        val_accuracy = accuracy(X_test, y_test)
        print(f"Validation Accuracy: {val_accuracy:.2f}%")
    model.train()
```

```python
def accuracy(X, y):
  correct = 0
  total = len(y)
  model.eval()

  with torch.no_grad():
    outputs = model(X)
    _, predicted = torch.max(outputs.data, 1)
    correct += (predicted == y).sum().item()

  return (correct/total)*100
```
